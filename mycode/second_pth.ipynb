```python
import torch

path = "./second_7862.pth"
checkpoint = torch.load(path, map_location='cpu')
print("文件类型:", type(checkpoint))


print("=== 字典顶层所有键 ===")
for idx, key in enumerate(checkpoint.keys()):
    print(f"{idx+1}. {key} → 类型：{type(checkpoint[key])}")

print(f"\n字典顶层键总数:{len(checkpoint.keys())}")
```

```python
from collections import OrderedDict

print("=== 训练进度信息 ===")
print(f"训练轮数(epoch):{checkpoint['epoch']}")
print(f"迭代步数(it):{checkpoint['it']}")

model_state = checkpoint["model_state"]  
print("\n=== 模型参数(model_state)详情 ===")
param_keys = list(model_state.keys())  
print(f"\n模型参数层总数:{len(param_keys)}")
for idx, key in enumerate(param_keys[:163]):
    print(f"  {idx+1}. {key}")
if len(param_keys) > 163:
    print(f"  ... 共{len(param_keys)}层,仅展示前163个")
```

查看所有层的张量形状

```python
for idx, key in enumerate(param_keys[:163]):
    print(f"  {idx+1}. {key}")
    print(f"    参数形状: {model_state[key].shape}, 数据类型: {model_state[key].dtype}")
    print(f"    该层参数总数：{model_state[key].numel():}")
if len(param_keys) > 163:
    print(f"  ... 共{len(param_keys)}层,仅展示前163个")
```

```python
first_key = param_keys[0]
first_param = model_state[first_key]
print(f"\n=== 示例1:查看「{first_key}」参数 ===")
print(f"  形状：{first_param.shape}")          # 张量形状（如卷积层：out_ch, in_ch, kH, kW）
print(f"  数据类型：{first_param.dtype}")      # 如torch.float32
print(f"  数值范围：{first_param.min():.4f} ~ {first_param.max():.4f}")  # 数值分布
print(f"  该层参数总数：{first_param.numel():,}")  # 该层参数个数
```

```python
target_key = "backbone_3d.conv_input.0.weight"
target_param = model_state[target_key]
print(f"\n=== 示例2:查看「{target_key}」参数 ===")
print(f"  数据类型：{target_param.dtype}") 
print(f"  形状：{target_param.shape}")
print(f"  前5个数值:{target_param.flatten()[:5]}")  
```
